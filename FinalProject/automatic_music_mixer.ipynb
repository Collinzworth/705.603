### Feature Selection
The amount of features that can be obtained from audio signals is extremely large. Signals can be represented in several different forms. For the case of time series signals, there is both the time domain and frequency domain representations. The time domain representation is the more well known of the two where the signal is represented as amplitude vs time. Amplitude of sound is typically measured in decibels. The frequency domain represention is more complicated but is able to extract a lot of information from the signal that the time domain can not easily identified. The frequency domain is found by taking the Fourier transformation of the signal. Different variants exist depending on which musical feature you are interested in. 

#### Chromagrams

Chromagrams are a spectral transformation that measure intensity and pitch. Chromagrams can help detect what pitch a sound is at and how intense. This helps with identifying music characteristics because it can detect which key a track is in and find synergetic sounds between two tracks. (8) provides research on training a model using chromagrams to detect if a patient's cough is a result of COVID-19, thus reducing invasive methods of detecting COVID-19 that were present at the time. The applications of the information chromograms reveal have a wide range of possibilities. (9) delves deeper into the mathematical importance of chromagrams and applies to to a chord recognition algorihtm. Chromagrams take care of the pitch aspect of musical information, but more is needed in order to be able to classify a piece of music more thoroughly.

#### Mel Frequency Cepstral Coefficients

(1) used Mel Spectrograms as features for music classification tasks. Mel-Frequency Cepstral Coefficients (MFCCS) are used alongside mel spectrograms to analyze and describe the human's perception on hearing noise. They were originally developed to characterize vocal excitation from the vocal tract (6). The mel frequency domain is a special transformation from the frequency domain in which another transformation is computed on the frequency domain. The result is a whole new domain known as "ceptral" vice "spectral". MFCCS are coefficients that describe the Mel Frequency domain. The coefficients aid in classifying the timbral aspects of the audio signal. Timbre refers to a sound's pitch and intensity. The way we interperet sound is related to the intensity and pitch of the sound, which is a complex non-linear relationship. MFCCs as a result are an appropriate feature for training audio models because of their ability to scale intensity and pitch to a human's perception.


#### Tempogram Ratio

(2) Grosche et al developed methods of extracting tempo information at a mid level representation. They develop methods to encode local tempo information. They derive the concept of cyclic tempograms which are useful in gaining insight about rythm components of the audio signal. Tempogram ratios are compact form of a tempogram that summarize energy at important multiples of the tempos (7). Since tempogram ratios encode the tempogram information in less dimensions, the tempogram ratio is used instead of the tempogram.


#### Feature Aggregation

With chromagrams, MFCCS, and tempogram ratios, three large portions of music information can be classified using a small amount of features. It is important to note that when a transformation is made on a audio file, a set of frames containing a vector for each feature are present. The features used in the algorithm are the mean and standard deviations of each of the frames yielded from the transformation. This allows the features to be put into a form that is low enough in dimensions and consistant among all transformed audio files. Since KNN models are sensitive to features the aggregated features are normalized between 0 and 1.

### Model Selection

#### K-Nearest Neighbor

K-Nearest Neighbor was chosen as the model for it's ability to find similar datapoints given a feature set. K-Nearest Neighbor is simple yet effective. However, K-Nearest Neighbor is influenced heavily by the number of features in a dataset. It is also influenced by magnitudes of the features. Because of these two attributes, PCA was implemented for dimensionality reduction and normalization was applied. When generating a new mix, the nearest neighbor that is not part of the current beat's artist or song is selected for the next beat to be mixed in.

Details on how the number of components were selected for PCA can be found in the *pca_dimensionality_reduction.ipynb* notebook.

**Note:** I attempted to use autoencoders for dimensionality reduction but ran into trouble with getting it to train on the large amounts of spectral audio data. ### Deployment

The deployment for this model would be pretty simple. It would require that a pod manages the server side library of trained data. Since KNN is a "lazy learner", the pod would have to manage a database that contains the trained data features. Potentially a user could upload a set of tracks to the database in order to query and get a mix based on their tracks. Depending on how the usage would go, serverless deployment could be used since the only time it would be used is when the user makes a query and requests a remix during the inference stage. If the number of remixes is low, then the serverless deployment option might be a good one. If the interactions are more complex and the music library is constantly being updated via tracks that users upload then it might be better to use a standard deployment strategy. Different pods could create different databases for users depending on their preferences. This would make the queries more tailored to their preferences which could be learned through a feedback system. This concept could take use of more recommendation systems by having individual pods containing their tailored database interact with eachother and update eachother based on a similarity measure of each user to one another. This way a user's database could be updated automatically with tracks that the system thinks the user would prefer. Between the user's feedback and user similarity the orchestration of the pods could assist with drift and prevent too much overfitting to a user's preference. Tailoring too much to a user's preference and not generalizing potentially new tracks to the list could lead to the program getting stale and the user getting disinterested. The ability to expand and incorporate new tracks based on similar user's preference would be a good way to combat this potentially negative consequence of this particular machine learning model.
