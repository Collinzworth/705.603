# PCA Dimensionality Reduction

In this notebook I will demonstrate the dimensionality reduction feature PCA performed on the dataset. The resulting number of dimensions after combining the various features and their aggregations came out to be 38 dimensions. 

### Check Explained Variance

```python
import os
from visualization.plot_pca import plot_pca

relative_filepath = os.getcwd() + "\\generated_models\\dataset\\dataset.pkl"

num_components = 38
plot_pca(num_components, relative_filepath)
```

From graph above, I tried 10, 15, 20 components with 20 yielding the best sounding results. 10 and 15 components resulted into much loss of information resulting in what appeared to be discontinuity involving the MFFCs. I could tell the loss of information was comming from the loss of MFCC information since the timbre continuity was not great when PCA reduced the dataset down to 10/15 components. PCA was able to shrink the size of the model from ~400 mb to ~200 mb, resulting in about 1/2 the original size which makes sense since we reduced from 38 columns to 20. 

The ability for pca to reduce the dataset size and maintain information allows the search to be more efficient and saves on disk space for the model. If the model became larger as it would in a production enviroment. Since a lot of the relationships in signal and sound processing are non-linear, I imagine that a non-linear method for dimensionality reduction would be a better option. PCA produced genuinely good results anyways though.
