## Genetic Algorithm Demonstration

This demonstration will show the genetic algorithm optimizing a randomly generated bipartite graph. Each improvement of the fitness throughout the generations will be plotted. Finally, the best solution will be compared with the original generated bipartite graph.

```python
import pylab as pl
import matplotlib.pyplot as plt
import Optimize_Layout.Gen_Bipartite_Graph as gbg
import Optimize_Layout.Optimize_Line_Crossings as olc
import Optimize_Layout.GA_Visualization as gv
import Optimize_Layout.Bi_Graph as big
import Optimize_Layout.Population as pop
```

```python
# Set Font Parameters
size=15
params = {'legend.fontsize': 'large',
          'figure.figsize': (20,8),
          'axes.labelsize': size,
          'axes.titlesize': size,
          'xtick.labelsize': size*0.75,
          'ytick.labelsize': size*0.75,
          'axes.titlepad': 25}
plt.rcParams.update(params)
```

```python
genetic_args = {}
genetic_args["pop_size"] = 20
genetic_args["max_gen"] = 300
genetic_args["tourny_perc"] = 0.2
genetic_args["elite_perc"] =  0.1
genetic_args["crossover_rate"] = 1.0
genetic_args["mutation_rate"] = 0.2

plot_network = False

network = gbg.gen_bipartite_graph(20, 20, 20, 20, 20)
#network = gbg.gen_bipartite_graph(35, 40, 35, 40, 40)

bipartite_graph = big.Bi_Graph(network)
bipartite_graph.set_node_dict_attr()

best_individuals = olc.opt_line_crossings(bipartite_graph, plot_network, **genetic_args)
```

### Best Individuals Over Generations

Below the best individuals are plotted with their respective fitness at the generation in which the improvement was made to that fitness. 

```python
plt.rcParams['figure.figsize'] = [10, 8] # Change values to edit plot size

max_x, max_y = pop.get_limits(best_individuals)
for frame in best_individuals:
    fig, ax = pl.subplots()
    diagram_to_plot = frame[0][1]
    fig, ax = gv.plot_segments(fig, ax, diagram_to_plot, max_x, max_y, frame[1])
    plt.draw()
    plt.pause(.01)
```

### Fitness Vs Generations

```python
plt.rcParams['figure.figsize'] = [24, 12]
gv.plot_fitness_evolution(best_individuals)
```

The above plot shows the convergence of the model around 150 generations. There generations where you see the fitness improvements stagnate comparatively the algorithm gets stuck for a few generations. The algorithm makes sudden jumps and the fitness starts to improve gain gradually. This is demonstrating the ability for the algorithm to "figure" things out and begin a new phase of improvements.

```python
plt.rcParams['figure.figsize'] = [24, 12]
best_individual = best_individuals[-1]
gv.compare_plots(bipartite_graph, "Original Graph", best_individual[0][1], "Optimized Graph", -1)
```
